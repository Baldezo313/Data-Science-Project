*Etape 3: D√©velopper le mod√®le et l'entra√Æner*  
Pour d√©velopper et entra√Æner un mod√®le, il a fallu beaucoup de maths dans le chapitre 2 : Entre la Fonction Co√ªt, les d√©riv√©es, l‚Äôalgorithme de Gradient Descent‚Ä¶
Dans Sklearn, tout cela est d√©j√† fait pour vous !  
  
Il vous suffit de d√©finir une variable *model* depuis le g√©n√©rateur **SGDRegressor** en entrant le nombre d‚Äôit√©rations que le *Gradient Descent* doit effectuer ainsi que le *Learning Rate*.  
  
Une fois le mod√®le d√©fini, il vous faut **l‚Äôentra√Æner**. Pour cela, il suffit d‚Äôutiliser la fonction **fit.**  
  
Par exemple, entra√Ænons notre mod√®le sur **100** it√©rations avec un Learning rate de **0.0001** :

```python
#cr√©er notre model a la main. 
def modelm(X, theta):
    return X.dot(theta)
#modelm(X, theta)

# Afficher notre model
plt.scatter(x,y)
plt.plot(x, modelm(X,theta), c='red')
plt.show()

```

```python
# Notre Fonction Cout

def cost_function(X,y,theta):
    m = len(y)
    return 1/(2*m) * np.sum((modelm(X,theta)-y)**2)

cost_function(X, y, theta)
    
```

* Implementation du Gradient Descent

```python
def grad(X,y, theta):
    m = len(y)
    return 1/m*X.T.dot(modelm(X,theta)-y) 

def gradient_descent(X,y,theta,learning_rate, n_iterations):
    cost_history = np.zeros(n_iterations)
    theta_history = np.zeros((iterations, 2))
    for i in range(0,n_iterations):
        theta = theta-learning_rate*grad(X,y, theta)
        # On enregistre le co√ªt de notre model pour chaque iteration
        cost_history[i] = cost_function(X, y, theta)
        theta_history[i,:] = theta.T
        
    return theta, cost_history, theta_history
```

**Machine Learning!!**

```python
theta_final, cost_history, theta_history = gradient_descent(X,y,theta, learning_rate=0.01, n_iterations=1000)
theta_final
```

```python
predictions = modelm(X,theta_final)
plt.scatter(x,y)
plt.plot(x, predictions, c='r')
```

Trcer la courbe de minimisation

```python
plt.plot(range(1000), cost_history)
```

* Coefficient de determination $R^{2}$:  
Pour √©valuer la performation d'un mod√®le de r√©gression lin√©aire

$$ R^{2} =1 - \frac {\sum ( ùëì(ùë•) ‚àí ùë¶ )^{2}}{\sum (y-\overline {y} )^{2}} $$ 

```python
def coef_determination(y, pred):
    # residu de la somme des carr√©e
    u = ((y-pred)**2).sum()
    v = ((y-y.mean())**2).sum()
    return 1- u/v

coef_determination(y, predictions)
    
```

### Les probl√®mes de Classification  
On cours du 1er partie, on a parl√© des probl√®mes de classification, qui consiste par exemple √† classer un email en tant que 'spam' ou 'non spam'.  
  
Dans ce genre de probl√®me, on aura un Dataset contenant une variable target **y** pouvant prendre 2 valeurs seulement, par exemple 0 et 1.  
***  
* si y=0, alors l'email n'est pas un spam  
* si y=1, alors l'email est un spam

On dit √©galement que l'on a **2 classes**, c'est une classification **binaire**.  
   
Pour ces probl√®mes, on ajoute au mod√®le une **fronti√®re de d√©cision** qui permet de classer un email dans la classe 0 ou la classe 1.  
Par exemple:  
***  
si $f(x) \leq 0.5 $ alors y=0  
si $f(x) > 0.5 $ alors y=1 

### 4.a Le mod√®le de R√©gression logistique  
Pour les probl√®mes de classification binaire, un mod√®le lin√©aire $F=X.\theta$ ne convient pas.  
On va d√©velopper alors une nouvelle fonction pour les probl√®mes de classification binaire, c'est la **fonction logistique** (aussi appel√© fonction **sigmo√Øde** ou tout simplement sigma $\sigma$). Cette fonction a la particularit√© d'√™tre toujours comprise entre 0 et 1.  
  
Pour coller la fonction logistique sur un Dataset (X,y), on y fait passer le produit matriciel $X.\theta$ ce qui nous donne le **mod√®le** de Logistic Regression:  
$$\sigma(X.\theta) = \frac {1}{1+ e^{-X.\theta}}$$  
A partir de cette fonction, il est possible de d√©finir une **fronti√®re de d√©cision**. Typiquement, on d√©finit un seuil √† **0.5** comme ceci:  
***  
y=0 si $\sigma(X.\theta) < 0.5$  
y=1 si $\sigma(X.\theta) \geq 0.5$   
***  



####  Fonction Co√ªt associ√©e √† la R√©gression Logistique  
Pour la r√©gression lin√©aire, la Fonction Co√ªt $J(\theta)=\frac {1}{2m} \sum_{i=1}^{m} ( X.\theta ‚àí ùë¶ )^{2}$ donnait une courbe **convexe**(qui pr√©sente un unique minima). C'est ce qui fait que l'algorithme de Gradient Descent fonctionne.  
  
En revanche, utiliser cette fonction pour le mod√®le de Logistique ne donnera pas de courbe convexe (d√ª √† la non-lin√©airat√©) et l'algorithme de Gradient Descent se **bloquera** au premier minima rencontr√©, sans trouver le minimum **global**  
  
Il faut donc d√©velopper une nouvelle Fonction Co√ªt sp√©cialement pour la r√©gression Logistique. On utilise alors la fonction **logarithme** pour **transformer** la fonction sigma en fonction convexe en s√©parant les cas o√π y=1 du cas o√π y=0.  
  
   
* **Fonction Co√ªt dans le cas o√π y=1**  
Voici la Fonction Co√ªt que l'on utilise dans les cas o√π y=1 $$J(\theta) = -log(\sigma(X\theta))$$  
*EXPLICATION*:  
Si notre mod√®le pr√©dit $\sigma(x)=0$ alors que y=1, on doit p√©naliser la machine par une **grande** erreur(un grand co√ªt). La fonction logarithme permet de tracer cette courbe avec une propri√©t√© convexe, ce qui poussera le Gradient Descent √† trouver les **param√®tres** $\theta$ pour un co√ªt qui tend vers 0.

* **Fonction Co√ªt dans les cas o√π y=0**  
cette fois la fonction Co√ªt devient:  
$$J(\theta) = -log(1- \sigma(X\theta))$$  
*EXPLICATIONS:*  
Si notre mod√®le pr√©dit $\sigma(x)=1$ alors que y=0, on doit p√©naliser la machine par une **grande** erreur(un grand co√ªt). Cette fois **-log(1-0)** donne la m√™me courbe, invers√©e sur l'axe vertical. 

* **Fonction Co√ªt Compl√®te**  
Pour √©crire la Fonction C√¥ut en une seule √©quation, on utilise l'astuce de s√©parer les cas y=0 et y=1 avec une annulation:  
$$J(\theta) = \frac {-1}{m} \sum y*log(\sigma(X\theta)) + (1-y)*log(1- \sigma(X\theta))$$

**GRADIENT DESCENT POUR LA REGRESSION LOGISTIQUE**  
L‚Äôalgorithme de Gradient Descent s‚Äôapplique exactement de la **m√™me mani√®re** que pour la r√©gression lin√©aire. En plus, la d√©riv√©e de la Fonction Co√ªt est la m√™me aussi ! On a :  $$Gradient: \; \frac {\partial J(\theta)}{\partial \theta} = \frac {1}{m} \sum (\sigma(X.\theta)-y).X$$  
$$Gradient Descent: \; \theta = \theta - \alpha*\frac {\partial J(\theta)}{\partial \theta}$$


*Resumer de la R√©gression Logistique**  
* $$mod√®le: \; \sigma(X.\theta) = \frac {1}{1+ e^{-X.\theta}}$$  
* $$Fonction \; Co√ªt: \; J(\theta) = \frac {-1}{m} \sum y*log(\sigma(X\theta)) + (1-y)*log(1- \sigma(X\theta))$$  
* $$Gradient: \; \frac {\partial J(\theta)}{\partial \theta} = \frac {1}{m} X^{T}.(\sigma(X.\theta)-y)$$   
* $$Gradient \; Descent: \; \theta = \theta - \alpha*\frac {\partial J(\theta)}{\partial \theta}$$

### L'algorithme de Nearest Neighbour  
L'algo. de **Nearest Neighbour** (le voisin le plus proche) permet de resoudre des probl√®mes de classification √† **plusieurs classes** de fa√ßon simple et tr√®s efficace.  
  
*PROMENADE EN MONTAGNE*  

Vous partez vous promenez en montagne avec un ami. Avant de partir, il fait 30¬∞C et votre ami vous dit qu'il a chaud. Arriv√© en montagne, il fait d√©sormais 10¬∞C et votre ami vous dit qu'il a froid.  
En redecendant la vall√©e, il fait maintenant 15¬∞C, pensez-vous que votre ami aura froid ?  
15¬∞C √©tant plus proche de 10¬∞C(froid) que de 30¬∞C(chaud), il semble l√©gitime de pr√©dire que votre ami aura froid.  
  
Voila l'essentielle de ce qu'il y √† savoir sur l'algo. **Nearest Neighbour**. Quand vous devez faire une nouvelle pr√©diction, trouvez dans dans votre **Dataset** l'exemple le plus **proche** par rapport aux **conditions** dans lesquelles vous √™tes.  
  
Cet exemple nous montre au passage que la **variet√©** et la **quantit√©** de donn√©es dans notre Dataset est primordiale!! Si vous ne disposez que de 2 points: -20¬∞C=froid; 20¬∞C=chaud, alors nous pourrions conclure que 1¬∞C est une temp√©rature chaud... Pas s√ªr que √ßa plaise √† tout le monde!

**K-Nearest Neighbour (K-NN)**  
*LA DISTANCE LA PLUS COURTE*  
Par exemple on regarde un nuage de point. Et on veut determiner l'exemple le plus proche du point vert. si c'est un element de la classe rouge. Alors l'algorithme de K-NN calcule ainsi la **distance** entre le point vert et les autres point du Dataset et associe le point vert √† la classe dont l'exemple est le plus proche en terme de distance.  
  
Typiquement, o utilise la **distance euclidienne** mais d'autres m√©triques sont parfois utiles, comme la distance de **Manhattan** ou bien la distance **cosinus**.  
Et **Sklearn** a impl√©ment√© touts les √©quations pour nous.  
  
*LE NOMBRE DE VOISIN K*  
Pour limiter les probl√®mes li√©s au bruit (ce qu'on appelle **Over fitting**) on peut demander √† l'algo. de trouver les **K** voisins les plus proches du point vert.  
cela **am√©liore** la qualit√© du mod√®le car il devient moins sensible aux impuret√©s et cas particuliers qui viendraient emp√™cher la bonne **g√©n√©ralisation**.(on verra √ßa dans la partie 7).

### Introduction aux R√©seaux de Neurones  
Les R√©seaux de Neurones sont des mod√®les bien plus **compexes** que tous les autres mod√®les de Machine Learning dans le sens o√π ils repr√©sentent des fonctions math√©matiques avec des **million** de coefficient(les param√®tres).

Pour d√©passer ces challenges, les chercheurs dans le domaine ont d√©velopp√©s des **variables** du Gradient Descent ainsi que d'autres techniques pour calculer plus rapidement les d√©riv√©es sur des millions de donn√©es. Parmi ces solutions on trouve:

* **Distributed Deep Learning**: Utilisation du **Cloud** pour **diviser** le traavail et le confier √† plusieurs machines.  
***  
  
*COMPRENDRE LES R√©seaux de Neurones*  
Sans plus tarder, voyons l'natomie d'un R√©seau de Neurones pour d√©mystifier ce concept.  
Il est constitu√© par un niveau d'entr√©es(*input layer*) √† gauche, un niveau de sorties(*output layer*) √† droite, et plusiseurs niveaux **cach√©s** entre les deux; des petits ronds qui sont appel√©s les **neurones** et qui repr√©sentent des **fonctions d'activation**. Pour un r√©seau de neurone basique, la **fonction Logistique** est utilis√© comme fonction d'activation(c'est le m√™me fonction qu'on a vu dans le partie4).  
  
Commen√ßons par analyser ce qui se passe dans **1 neurone**  
  
* **R√©seau de Neurone √† 1 Neurone: Le perceptron**  
Le r√©seau de Neurones le plus simple qui existe porte le nom de perceptron. Il est **identique** √† la R√©gression Logistique.  
  
Les entr√©es du neurone sont les **features x** multipli√©es par des **param√®tres** $\theta$ √† apprendre. Le calcul effectu√© par le neurone peut √™tre divis√© en deux √©tapes:  
***  
* 1. Le neurone calcule la **somme z** de tous les entr√©es $z(x.\theta) = \sum x.\theta $. C'est un calcul lin√©aire  
* 2. Le neurone passe **z** dans sa fonction d'activation. Ici la fonction sigmo√Øde (fonction Logistique). C'est un calcul non-lineair. $y=a(z)= \frac {1}{1+e^{-z}}$  
  
***  
*Note*: On utilise souvent d'autres fonctios d'activation que la fonction sigmo√Øde pour simplifier le calcul du gradient et ainsi obtenir des cycles d'apprentissage **plus rapides**: La fonction tangente hyperbolique *tanh(z)*, la fonction *Relu(z)*.  
  
* **R√©seaux √† plusieurs neurones: le Deep Learning**  
Pour cr√©er un r√©seaux de Neurones, il suffit de d√©velopper plusieurs de ces perceptrons et de les **connecter** les uns aux autres d'une fa√ßon particuli√®re:  
***  
* On r√©unit les neurones en **colonne** (on dit qu'on les r√©unit en **couche**, en layer). Au sein de leur colonne, les neurones ne sont pas connect√©s entre eux.  
* On connecte toutes les **sorties** des neurones d'une colonne √† gauche aux **entr√©es** de tous les neurones de la colonne de droite qui suit.  
***  
On peut ainsi construire un r√©seau avec autant de couche et de neurones que l'on veut. Plus il y a de couches, plus on dit que le r√©seau est **profond** (*deep*) et plus le mod√®le devient **riche**, mais aussi **difficile** √† entra√Æner. C'est √ßa, le **Deep Learning**.  
  
Dans les d√©tails, un r√©seau plus simple(√† 3 neurones) nous donnerait la sortie $a_{3}=\sigma(\theta_{1}a_{1}+\theta_{2}a_{2}) \; o√π \; \theta_{1} \; et \; \theta_{2}$ sont les coefficients li√©s aux connections entre neurones $a_{1} -> a_{3} \; et \; a_{2} -> a_{3}$. Ce sont les **param√®tres** de notre mod√®le.  
  
* **C‚Äôest comme le cerveau humain ?**  
On a longtemps fait le rapprochement entre le cerveau humain et les Neural Network pour d√©montrer la puissance de ces algorithmes. Voici l‚Äôanalogie qui est encore aujourd‚Äôhui pr√©sent√©e aux novices :  
  
La fonction d‚Äôactivation produit une sortie si les entr√©es qu‚Äôelle re√ßoit d√©passent un certain seuil, √† la mani√®re qu‚Äôun neurone biologique produit un signal √©lectrique en fonction des stimulus qu‚Äôil re√ßoit aux Dendrites (ce sont les entr√©es du neurone).  
  
Dans un Neurone, ce signal circule jusqu‚Äôaux diff√©rents terminaux de l‚Äôaxone pour √™tre transmis √† d‚Äôautres neurones, tout comme la fonction activation envoie sa sortie aux neurones du niveau suivant.  
  
En r√©alit√©, les R√©seaux de Neurones n‚Äôont rien √† voir avec le cerveau humain.  
Un R√©seau de Neurone n‚Äôest en fait qu‚Äôune √©norme compos√©e de milliers de fonctions math√©matiques, et aujourd‚Äôhui les neuroscientifiques ont d√©montr√© que le fonctionnement du cerveau d√©passe de loin l‚Äôarchitecture ¬´ simpliste ¬ª des r√©seaux de neurones.  
  
**Pourquoi** utilise-t-on alors l‚Äôexpression de R√©seau de Neurone ? * **L'entr√Ænement d'un R√©seau de Neurone**  
Rappelez-vous : Pour r√©soudre un probl√®me de Supervised Learning, il vous faut les 4 √©l√©ments suivants.  
***  
* 1. Un Dataset  
* 2. Un Mod√®le et ses param√®tres  
* 3. Une Fonction Co√ªt et son gradient  
* 4. Un Algorithme de minimisation (Gradient Descent)  
***  
* Pour le **Dataset**, pas de probl√®me, il suffit de disposer d'un tableau **(X,y)** comme pour les autres probl√®mes. Les features $(x_{1},x_{2}, x_{3}...)$ sont distribu√©es √† l'entr√©e du r√©seau (dans le premier *layer*)  
  
* Pour programmer le **mod√®le**, il faut embo√Æter les fonctions des diff√©rent niveaux d'activation les unes dans les autres. C'est ce qu'on appelle **Forward Propagation** (faire le chemin des entr√©es X vers la sortie y).  
  
* Pour exprimer la Fonction Co√ªt et son gradient, c'est math√©matiquement d√©licat. Il faut calculer la **contribution** de chaque neurone dans **l'erreur** finale. Tout ce qu'on peut dire c'est que cela est possible avec une technique appel√©e **Back Propagation** (faire le chemin dans le sens inverse: y vers X).  
  
* Enfin, pour minimiser la Fonction Co√ªt, il suffit d'utiliser Gradient Descent en utilisant les gradient calcul√©es avec Back Propagation. Le gradient Descent en lui-m√™me n'est **pas diff√©rent** de celui que nous avons vu pr√©c√©dement.  





### Programmer notre premier R√©seau de Neurones pour identifier des esp√®ces d'Iris.  
Normalement, on d√©veloppe des R√©seaux de Neurones avec un Framework comme Tensorflow. Cependant, on va montrer comment le faire avec **Sklearn**. Pour cela, il faudra importer *MLPClassier*(qui signifie: Multi-Layer Perceptron Classifier).  
  
Cette fois, on va d√©velopper un programme capable de reconnaitre une esp√®ce parmi plusieurs de la famille des Iris. L'algo. utilise 4 *features* pour effectuer son calcul:  
***  
‚Ä¢ ùíôùüè : La longueur du p√©tale  
‚Ä¢ ùíôùüê : La largeur du p√©tale  
‚Ä¢ ùíôùüë : La longueur du s√©pale  
‚Ä¢ ùíôùüí : La largeur du s√©pale  
***  


```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
```

* Charger les donn√©es

```python
iris = load_iris()

Xi = iris.data
yi = iris.target

Xi.shape
```

Notre Dataset comprend 150 exemples et 4 variables

* Visualisation des donn√©es

```python
colormap = np.array(['Red', 'green', 'blue'])
plt.scatter(Xi[0:100, 3], Xi[0:100, 1], c=colormap[y])
plt.show()
```

Pour d√©velopper un r√©seau √† 3 *hidden layers* et 10 neurones dans chaque *layer*, on utilise le code suivant:*hidden_layer_sizes=(10,10,10)*  


* Cr√©ation du mod√®le

```python
modelneu = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)
modelneu.fit(Xi, yi)
modelneu.score(Xi, yi)
```

On a obtenu un score de 98% pour notre Neural Network. Cela signifie que la machine arrive √† pr√©dire la bonne esp√®ce de fleur √† 98% du temps

## 6. Apprentissage Non-Supervis√©  
Dans cette deuxieme partie de machine learning, on va apprendre l'algo. le plus populaire dans cette partie: Le **K-Mean Clustering** , qui permet de segmenter des donn√©es clients dans le monde du marketing, ou bien de faciliter la recherche scientifique en associant ensemble des mol√©cules/materiaux/ph√©nom√®nes semblables.

### Unsupervised Learning  
* **Le Probl√®me de l'Apprentissage Supervis√©**  
D'un certain point de vue, l'apprentissage supervis√© consiste √† enseigner √† la machine des choses que nous **connaissons d√©j√†**, √©tant donn√© que nous construisons √† l'avance un Dataset qui contient des questions **X** et des r√©ponses **y**.  
Maintenant que faire si on dispose d'un Dataset sans valeur **y** ? Que faire si on veut que la machine nous aide √† compl√©ter nos connaissances en apprenant certaines chose que nous ignorons ?   
  
Rappelons-nous, dans la premi√®re chapitre, nous avons parl√© d'apprendre la langue chinoise √† partir d'un livre de traduction (x,y). Que faire alors si on vous retire le bouquin et qu'on vous envoie vivre en chine ?  
Arrivez-vous √† apprendre le chinois tout seul ? Il y a forcement un moyen d'y parvenir, c'est ce qu'on appelle l'apprentissage non-supervis√© (**Unsupervised Learning**).  
* **Comment apprendre sans exemple de ce qu'il faut apprendre ?**  

Dans l'apprentissage non-supervis√©, on dispose ainsi d'un Dataset (x) sans valeur (y), et la machine **apprend** √† reconnaitre des **structures** dans les donn√©es (x) qu'on lui montre.  
  
On peut ainsi regrouper des donn√©es dans des clusters (c'est le **Clustering**), d√©tecter des **anomalies**, ou encore **r√©duire** la **dimension** de donn√©es tr√®s  riches en compilant les dimensions ensembles.

### Algorithme de K-Mean Clustering  
Le *K-Mean Clustering* est sans doute l'algo. le plus populaire pour les probl√®mes de **Clustering** (regrouper des donn√©es selon leur structure commune). Il est souvent utilis√© en marketing pour cibler des groupes de clients semblables pour certaines campagnes publicitaires.  
  
L'algo fonctionne en 2 √©tapes r√©p√©t√©es en boucle. On commence par placer au hasard un nombre **K** de **centres** dans notre nuage de points. Ensuite:  
L'√©tape 1 consiste √† rallier chaque exemple au centre le **plus proche**. Apr√®s cette √©tape, nous avons **K Clusters.** L'√©tape 2 consiste √† d√©placer les centres **au milieu** de leur Cluster.  
On r√©p√®te ainsi les √©tapes 1 et 2 en boucle jusqu'√† ce que les centres ne **bougent plus**

#### Over fitting et R√©gularisation  
Dans les chapitres d'apprentissage spervis√©, nous avons d√©velopp√© des mod√®les en cherchant √† **minimiser** les erreurs avec le Dataset. On a affirm√© qu'un bon mod√®le nous donne de petites erreurs. En fait ce n'est pas si simple...  
On peut avoir deux mod√®le **A** et **B**. Et que le mod√®le **B** soit beaucoup mieux ajust√© autour des nuages points que le mod√®le **A** i.e ne donne **aucune erreur** par rapport au Dataset. Et pourtant le mod√®le **A** semble plus convaincant, alors que celui-ci donne une Fonction Co√ªt **plus √©lev√©e**.  
En fait dans ce genre de cas, le mod√®le **B** souffre d'un probl√®me appel√© **Over fitting** qui est un ph√©nom√®ne **tr√®s courant** en Machine Learning et qu'il faut **absolument √©viter**.  
  
* **Over fitting: A vouloir aller trop loin, on va TROP loin**  
On parle d'*Over fitting* pour dire que le **mod√®le** s'est **trop sp√©cialis√©s sur les donn√©es** qui lui ont √©t√© fournies et a **perdu** tous sens de **g√©n√©ralisation**.  
  
Un *Over fitting* survient le plus souvent quand un mod√®le **trop complexe** (avec trop de param√®tres ou trop de features) √† √©t√© entra√Æn√©. Dans ce cas, le mod√®le a certe un faible co√ªt $J(\theta)$, mais il a aussi ce qu'on appelle une **grande variance**. Consequence: Un mod√®le moins performant que pr√©vu quand on le soumet √† de nouvelles donn√©es et une machine qui confond les pommes et les poires.  
  
On pourrait alors se dire qu'il suffit de d√©velopper des mod√®les moins complexes avec moins de *features...* Et pouf ! plus de probl√®me de variance!!  
  
C'est vrai, mais on risque alors d'avoir un mod√®le erron√© qui manque de pr√©cision. On appelle √ßa **Under fitting**, et on dit que le mod√®le a un **grand biais**.  
Ce probl√®me touche √† la fois les r√©gressions et les classifications:  
  
Comment trouver alors le **juste milieu** entre biais et variance ? C‚Äôest une des grandes questions √† laquelle sont confront√©s les Data Scientists.  
Il existe une m√©thode qui permet de garder toutes les features d'un mod√®le tout en **r√©gulant l'amplitude** des param√®tres $\theta$. Cette m√©thode porte le nom de **R√©gularisation**.  
  
* **La R√©gularisation**  
Elle permet de limiter la variance d'un mod√®le sans sacrifier son biais. Pour cela, diff√©rentes techniques existent:  
***  
1. On peut l√©g√®rement **p√©naliser** la Fonction Co√ªt du mod√®le en **ajoutant** un `terme de p√©nalit√©` sur ses param√®tres. Pour la r√©gression lin√©aire, la Fonction Co√ªt devient alors:  
$$J(\theta)=\frac {1}{2m} \sum (F(X)-Y)^{2}+\lambda \sum \theta^{2} \; (Ridge \; ou \; L2 \; R√©gularisation)$$  
Le facteur de r√©gularisation $\lambda$ correspond au niveau de p√©nalit√©: s'il est trop grand, on risque l*Under fitting*, et s'il est trop faible, c'est l'*Over fitting*. On peut le contr√¥ler directement dans **Sklearn**.  
2. Pour le K-Nearest Neighbour, on peut augmenter la **valeur de K** (nbr de voisins) Le mod√®le ne tient alors pas compte des anomalies noy√©es dans la masse.  
3. Pour les r√©seaux de Neurones, un technique nomm√©e **Dropout** p√©nalise le mod√®le en d√©sactivant al√©atoirement certains neurones √† chaque cycle de Gradient Descent. Le R√©seau perd alors l√©g√®rement de ses facult√©s et est moins sensible aux d√©tails.

***  
Essayons de voire ce que √ßa donne la r√©gression lin√©aire classique  en d√©veloppant un mod√®le polyn√¥miale de degr√©s 10.  

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

* Creation d'un Dataset x, y

```python
np.random.seed(0)

x10 = np.linspace(0,5,10)
y10 = x10-2*(x10**2)+0.5*(x10**3)+np.random.normal(-2,2, 10)
plt.scatter(x10,y10)
plt.show()
```

* Creation de plusieurs features pour notre mod√®les

```python
X10 = x10[:, np.newaxis]
X10 = PolynomialFeatures(degree=10,include_bias=False).fit_transform(X10)
X10.shape
```

Entra√Ænement du mod√®le. Ici on utilise les Equations Normales (LinearRegression). Les Equations Normales reposent sur la m√©thode des moindre car√©es, c'est plus rapide que le Gradient Descent.  


```python
model10 = LinearRegression()
model10.fit(X10,y10)
print('Coeff R2=', model10.score(X10,y10))

plt.scatter(x10, y10, marker='o')
plt.plot(x10, model10.predict(X10), c='red')
plt.show()
```

Dans Sklearn, on peut d√©velopper un mod√®le avec r√©gularisation gr√¢ce au mod√®le **Ridge**: *Sklearn.linear_model.Rigde*. Le mod√®le a certe un coefficient $R^{2}$ plus faible, mais il produit une meilleur g√©n√©ralisation et fera donc moins d'erreurs sur les donn√©es futures.  

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1)    # alpha est le facteur de r√©gularisation
ridge.fit(X10,y10)
print('COeff R2=', ridge.score(X10,y10))

plt.scatter(x10, y10, marker='o')
plt.plot(x10, ridge.predict(X10), c='green')
plt.show()
```

On sait maintenant que la performance **r√©elle** d'un  mod√®le de M.L ne repose pas simplement sur sa Fonction Co√ªt:  
On risque l*Over fitting* chaque fois qu'un mod√®le se **sp√©cialise trop** sur les donn√©es qu'on lui donne √† √©tudier.  
  
Mais comment √™tre s√ªr de la performance que notre mod√®le aura sur des donn√©es **futures**, i.e des donn√©es sur lesquelles iln'aura **pas √©t√© entra√Æn√©** ?  
La r√©ponse est dans la question ! Il faut entra√Æner notre mod√®le sur **une partie seulement** du Dataset et utiliser la seconde partie pour √©valuer la vraie performance de notre mod√®le. On appelle cela le **Train set** et le **Test set**  
* **Diagnostiquer un mod√®le de Machine Learning**  
*Train set et Test set*  
La bonne mani√®re de mesurer la performance de notre mod√®le de M.L est de tester celui-ci sur des donn√©es qui n'ont pas servi √† l'entra√Ænement. On divise ainsi le Dataset **al√©atoirement** en deux parties avec un rapport 80/20:  
***  
* **Train set (80%)**, qui permet √† la machine d'entra√Æner un mod√®le.  
* **Test set (20%)**, qui permet d'√©valuer la performance du mod√®le.  
***  
  
Pour cr√©er un *Train set* et *Test set* √† partir de notre Dataset, on utilise la fonction **train_test_split** de Sklearn:  

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

* Cr√©ation d'un Dataset Aleatoire

```python
np.random.seed(0)
xd, yd = make_regression (n_samples=100, n_features=1, noise=10)
yd = np.abs(yd)+ yd+ np.random.normal(-5,5,100)
plt.scatter(xd,yd)
```

* Creation des Train_set et Test_set √† partir du Dataset

```python
x_train, x_test, y_train, y_test = train_test_split(xd,yd, test_size=0.2)
```

* Visualisation des Train_set et Test_set


```python
plt.scatter(x_train, y_train, c='blue', label='Train_set')
plt.scatter(x_test, y_test, c='red', label='Test_set')
plt.legend()
plt.show()
```

On peut ensuite entr√Æner notre mod√®le sur le $TrainSet:(x_{train}, y_{train})$, puis l'√©valuer sur le $TestSet:(x_{test},y_{test})$

Dans le code ci-dessous on a d√©cid√© de mettre en pratique tout ce que nous avons vu:  
On cr√©e plusieurs *features polynomiales* (d¬∞ 10) et on utilise la r√©gularisation de Ridge pour √©viter l*Over Fitting*.

```python
from sklearn.linear_model import Ridge



X= PolynomialFeatures(degree=10,include_bias=False).fit_transform(xd)
x_train, x_test, y_train, y_test = train_test_split(X, yd, test_size=0.2)

model = Ridge(alpha=0.1, random_state=0)
model.fit(x_train, y_train)

print('Coeff R2 sur Train_test:', model.score(x_train, y_train))
print('Coeff R2 sur Test_test:', model.score(x_test, y_test))

plt.figure(figsize=(8,6))
plt.scatter(xd, yd, c='blue')

a = np.linspace(-2,2,100).reshape(100,1)
A = PolynomialFeatures(degree=10, include_bias=False).fit_transform(a)
plt.plot(a, model.predict(A), c='green', lw=2)
plt.show()
```

On Obtient un score de **93%** pour l'entra√Ænement et un score de **78%** pour l'√©valuation sur des donn√©es nouvelles

* **Rep√©rer un probl√®me d'Under fitting ou d'Over fitting**  
La technique la plus efficace pour rep√©rer si notre mod√®le a un probl√®me de biais(*Under fitting*) ou de variance (*Over fitting*) consiste √† analyser les **erreurs** (la Fonction Co√ªt) sur le *Train set* $J(\theta)_{train}$ et le *Test set* $J(\theta)_{test}$:  
***  
* Si les **erreurs sont grandes** sur le *Train set* et le *Test Set*, alors le mod√®le a un grand **biais**, et il faut d√©velopper un mod√®le plus complexe ou d√©velopper plus de features.  
* Si les erreurs sont **faibles** sur le *Train Set*, mais sont **grandes** sur le *Test Set*, alors le mod√®le a une grande variance.  
  
* **Que faire en cas d'Over fitting ou Under fitting ?**  
Dans le cas o√π notre mod√®le a un grand biais (*Under fitting*) on peut:  
***  
* Cr√©er un mod√®le plus complexe, avec plus de param√®tres.  
* Cr√©er plus de *feature* √† partir des *features* existantes.  
* Diminuer le *Learning Rate du Gradient Descent* (si le *Learning Rate* est trop grand, la Fonction Co√ªt ne converge pas)  
* R√©colter plus de *features* dans les donn√©es (parfois une *feature* importante n'a pas √©t√© r√©colt√©e)  
  
***  
A l'inverse, si notre mod√®le a une grande variance (*Over fitting*) on peut:  
  
***  
* Utiliser la r√©gularisation  
* Utiliser un mod√®le avec moins de param√®tres ou un Dataset avec moins de *features*  
* Collecter plus de donn√©es (avoir un Dataset plus grand permet de d√©velopper un mod√®le qui g√©n√©ralise mieux)
