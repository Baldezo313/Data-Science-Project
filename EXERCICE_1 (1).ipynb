On s'intéresse à un "melange gaussien".  
On considère un n-échantillon d'une V.A. X de densité:  
$$f_{a,p}(x) = p \phi(x-a) + (1-p) \phi (x+a)$$,  
où $(a,p) \in R^{+*}x[0,1]$, et $\phi$ désigne la densité de la loi $N(0,1): \phi (x) = \frac {1}{\sqrt {2 \pi)}} e^{- \frac {x^2}{2}}$  
   
Nous avons vu qu'on pouvait estimer *a* et *p* par la méthode des moments, en posant:  
 $\hat a  = \sqrt {|\hat m_2 -1|}$ et $\hat p = \frac {\hat m_1 + \hat a}{2 \hat a}$  
(Avec $\hat m_i = \frac {1}{n} \sum _{k=1}^n X_k^i)$ .Ces estimateurs sont biaisés, mais fortement consistants.

**Question 2**  
Réaliser des n-échantillon de la loi $f_{a, p}$ pour n = 200,2000 et 20000.

```python


import numpy as np
import numpy.random as npr
np.random.seed(1000)

# pour n = 200
fnormal_200 =npr.normal(0,1,200)+(3*npr.binomial(1, 0.5, 200) -1)*2
fnormal_200 
```

```python
# pour n = 2000
fnormal_2000 =npr.normal(0,1,2000)+(3*npr.binomial(1, 0.5, 2000) -1)*2
fnormal_2000 
```

```python
# pour n = 20000
fnormal_20000 =npr.normal(0,1,20000)+(3*npr.binomial(1, 0.5, 20000) -1)*2
fnormal_20000 
```

**Question 3**  
Pour chacun des valeurs de n c_dessus, comparons les valeurs de $\hat a$ et *a* $\hat p$ et *p*. 

```python
# Pour n = 200
m_1 = np.mean(fnormal_200)      # m1 chapeau
m_2 = np.mean(fnormal_200*fnormal_200)     # m2 chapeau

a_estime = np.sqrt(np.abs(m_2 -1))
p_estime = (m_1 + a_estime) / (2*a_estime)

print("estimateur de a=" ,a_estime)
print('estimateur de p = ',p_estime)
```

```python
# Pour n = 2000
m_1 = np.mean(fnormal_2000)      # m1 chapeau
m_2 = np.mean(fnormal_2000*fnormal_2000)     # m2 chapeau

a_estime = np.sqrt(np.abs(m_2 -1))
p_estime = (m_1 + a_estime) /(2*a_estime)

print("estimateur de a=" ,a_estime)
print('estimateur de p = ',p_estime)
```

```python
# Pour n = 20000
m_1 = np.mean(fnormal_20000)      # m1 chapeau
m_2 = np.mean(fnormal_20000*fnormal_20000)     # m2 chapeau

a_estime = np.sqrt(np.abs(m_2 -1))
p_estime = (m_1 + a_estime) / (2*a_estime)

print("estimateur de a=" ,a_estime)
print('estimateur de p = ',p_estime)
```

On constate que pour chaque valeurs de *n* ci-dessus, les valeurs des estimateurs $\hat a$ et $\hat b$ sont superieur aux valeurs réelles de *a* et *p*.    
   
Oui quand on continu a augementer la valeur de *n* on observe bien des convergences pour ces deux estimateurs respectivement vers les valeurs de *a* et de *p*.  

**Question 4**  
Reproduire N = 1000 fois les estimations précédentes, On obtient donc 1000 estimations différentes pour a et p. 

```python


a_chap = []
p_chap = []

for n in range(1000):
  fnormal_n = npr.normal(0,1,n)+(3*npr.binomial(1, 0.5, n) -1)*2

  m_1 = np.mean(fnormal_n)      # m1 chapeau
  m_2 = np.mean(fnormal_n*fnormal_n)
  a_chap.append(np.sqrt(np.abs(m_2 -1)))
  
  p_chap.append((m_1 + a_estime) / (2*a_estime))


a_chap
```

```python
p_chap
```

**Question 5**  
Représenter, sur deux histogrammes distincts, les distributions des 1000 valeurs obtenues pour $\hat a$ (resp. pour $\hat p$. On fera figurer sur les histogrammes les vrais valeurs de a et de p.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,8))
plt.hist(a_chap, bins = 50, rwidth=0.9,color="b",edgecolor="red", label="estimation de a")
plt.axvline(x=a, color='g', linestyle='--', label = "vrai valeur de a")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(p_chap, bins = 50,edgecolor="green", label="estimation de p")
plt.axvline(x=p, color='g', linestyle='--', label = "vrai valeur de p")

plt.legend()
plt.show()
```

**Question 6:**  
Pour chacun des estimateurs $\hat a$ et $\hat p$, en utilisant ces 1000 valeurs  évaluer l'espérance et le biais, la variance et le risque quadratique.

```python

# Esperance de a chapeau
E_a_hat = np.nanmean(a_chap)

# variance de a chapeau
V_a_hat = (np.nanstd(a_chap))**2

# le biai de a chapeau
B_a_hat = E_a_hat - a

# Risque quadratique moyenne de a chapeau
R_a_hat = B_a_hat**2 + V_a_hat


print("Esperance de a chapeau est de:", E_a_hat)
print("variance de a chapeau est de :", V_a_hat)
print("le biai de a chapeau est de :", B_a_hat)
print("Risque quadratique moyenne de a chapeau est de :", R_a_hat)
```

```python

# Esperance de p chapeau
E_p_hat = np.nanmean(p_chap)

# variance de a chapeau
V_p_hat = (np.nanstd(p_chap))**2

# le biai de a chapeau
B_p_hat = E_p_hat - p

# Risque quadratique moyenne de a chapeau
R_p_hat = B_p_hat**2 + V_p_hat



print("Esperance de p chapeau est de:", E_p_hat)
print("variance de p chapeau est de :", V_p_hat)
print("le biai de p chapeau est de :", B_p_hat)
print("Risque quadratique moyenne de a chapeau est de :", R_p_hat)
```
